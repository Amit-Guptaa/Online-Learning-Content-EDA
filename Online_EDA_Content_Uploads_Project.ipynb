**CONTENT UPLOAD**

Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Data Fetching

```python
df = pd.read_csv("content_upload.csv")
```

Header

```python
print("Data Loaded Successfully\n")
print(df.head())
```

Date and Time Analysys

```python
df['upload_date'] = pd.to_datetime(df['upload_date'])
```

```python
df['month'] = df['upload_date'].dt.month
monthly_uploads = df.groupby('month').size()
```

Monthly Trends

```python
print("\nMonthly Uploads:")
print(monthly_uploads)
```

```python
category_count = df['category'].value_counts()
print("\nCategory-wise Uploads:")
print(category_count)
```

Plotting Graphs

```python
category_count.plot(kind='bar')
plt.title("Category-wise Content Uploads")
plt.xlabel("Category")
plt.ylabel("Number of Uploads")
plt.show()
```

```python
monthly_uploads.plot(kind='bar')
plt.title("Monthly Content Upload Trend")
plt.xlabel("Month")
plt.ylabel("Number of Uploads")
plt.tight_layout()
plt.show()
```
